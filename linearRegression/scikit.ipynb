```python
# Import Libraries

from sklearn import datasets
from sklearn.linear_model import LinearRegression

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Practice Dataset with Noise

X, y = datasets.make_regression(n_samples=100, n_features=1, noise=80)

plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
```

```python
# Feature and Regression Value

X, y
```

```python
# Number of Feature and regression value

print(type(X), X.shape)
print(type(y), y.shape)
```

```python
# Linear Regression fitting and prediction

reg = LinearRegression()

reg.fit(X, y)

predictions = reg.predict(X)
```

```python
# Variable for equation of a line

reg.coef_, reg.intercept_
```

```python
# Plot of predicted line vs feature

plt.scatter(X, y)
plt.plot(X, predictions, color='black', linewidth=2, label='Predicted Line')
plt.legend()
```
